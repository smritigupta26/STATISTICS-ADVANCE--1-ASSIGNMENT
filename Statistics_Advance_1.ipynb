### *Q1. What is a random variable in probability theory?*

ANS 1. A *random variable* is a numerical outcome of a random phenomenon. It assigns a numerical value to each outcome in a sample space.

* *Continuous random variable*: Takes an infinite number of values within a given range (e.g., height of students).

### *Q16. How do you calculate a z-score, and what does it represent?*

ANS 16.

$$

$$

Where:

* $X$ = value
* $\mu$ = mean
* $\sigma$ = standard deviation
  It represents how far a data point is from the mean in standard deviation units.

Q2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

low_bound = 1
high_bound = 7

outcomes = np.arange(low_bound, high_bound)

pmf_values = randint.pmf(outcomes, low=low_bound, high=high_bound)

plt.figure(figsize=(8, 6))
plt.plot(outcomes, pmf_values, 'bo', ms=8, label='PMF values')
plt.vlines(outcomes, 0, pmf_values, colors='b', lw=3, alpha=0.5, label='PMF bars')
plt.title('Probability Mass Function (PMF) of a Discrete Uniform Distribution')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.xticks(outcomes)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
import matplotlib.pyplot as plt

def bernoulli_pdf(p):
    x_vals = [0, 1]
    pdf_vals = [1 - p, p]
    return x_vals, pdf_vals

# Example with p = 0.6
x_vals, pdf_vals = bernoulli_pdf(0.6)

plt.bar(x_vals, pdf_vals, tick_label=x_vals)
plt.title("PDF of Bernoulli Distribution (p=0.6)")
plt.xlabel("x")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

Q11.Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
num_samples = 10000
sample_size = 30
original_distribution_mean = 5
original_distribution_std = 2

# Create a skewed distribution (e.g., exponential)
# You can change this to other distributions (uniform, etc.)
original_distribution = np.random.exponential(scale=original_distribution_mean, size=num_samples)

# Store sample means
sample_means = []

# Simulate multiple samples and calculate their means
for _ in range(num_samples):
    # Generate a sample from the original distribution
    sample = np.random.choice(original_distribution, size=sample_size, replace=True)
    # Calculate the mean of the sample
    sample_mean = np.mean(sample)
    # Store the sample mean
    sample_means.append(sample_mean)

# Plot the original distribution and the distribution of sample means
plt.figure(figsize=(12, 5))

# Original distribution
plt.subplot(1, 2, 1)
plt.hist(original_distribution, bins=50, density=True, alpha=0.6, color='blue', label='Original Distribution')
plt.title('Original Distribution (e.g., Exponential)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

# Distribution of sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=50, density=True, alpha=0.6, color='red', label='Sample Means')
plt.title('Distribution of Sample Means (CLT)')
plt.xlabel('Mean Value')
plt.ylabel('Density')

# Overlay a normal distribution for comparison
x = np.linspace(min(sample_means), max(sample_means), 100)
plt.plot(x, 1/(np.std(sample_means) * np.sqrt(2 * np.pi)) * np.exp( - (x - np.mean(sample_means))**2 / (2 * np.std(sample_means)**2) ),
         linewidth=2, color='green', label='Normal Distribution')
plt.legend()

plt.tight_layout()
plt.show()
```

Q16.Create a confidence interval for a dataset using Python and interpret the result.

```python
from scipy.stats import t

data = np.random.normal(50, 10, 30)
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)
confidence = 0.95

t_crit = t.ppf((1 + confidence) / 2, df=n-1)
margin_error = t_crit * (std / np.sqrt(n))
lower = mean - margin_error
upper = mean + margin_error

print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

Q17.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
from scipy.stats import t

# Generate sample data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=30)

# Sample statistics
mean = np.mean(data)
std_dev = np.std(data, ddof=1)  # Sample standard deviation
n = len(data)
confidence_level = 0.95

# t-critical value
t_crit = t.ppf((1 + confidence_level) / 2, df=n-1)

# Margin of error
margin_error = t_crit * (std_dev / np.sqrt(n))

# Confidence interval
lower_bound = mean - margin_error
upper_bound = mean + margin_error

print(f"Sample Mean: {mean:.2f}")
print(f"Confidence Interval (95%): ({lower_bound:.2f}, {upper_bound:.2f})")
```
